```python
from autoscraper import AutoScraper
#!pip install AutoScraper
import pandas as pd
```

```python
url = "https://www.pepperfry.com/site_product/search?q=table&src=table&as=0&autoSuggest=1"
# can be any url from any page
```

```python
list_1 = ["₹ 4,999","Genji Study Table In Columbia Walnut Finish","708"]
# copying features to be scraped in a list
```

```python
scraper = AutoScraper()
```

```python
result = scraper.build(url , list_1)
```

```python
print(result)
```

```python
scraper.get_result_similar(url , grouped = True)
```

```python
scraper.set_rule_aliases({'rule_6w9p':'Title','rule_z0al':'Price','rule_zcze':'Ratings'})
scraper.keep_rules(['rule_6w9p','rule_z0al','rule_zcze'])
scraper.save('Pepperfry-Search')
```

```python
results = scraper.get_result_similar(url,group_by_alias = True)
```

```python
results['Title']
```

```python
df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in results.items() ]))
df.to_csv('Pepperfry.csv')
# getting this result into a csv or excel file
```
